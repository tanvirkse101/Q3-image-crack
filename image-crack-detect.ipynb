## Libraries ##

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Get Image ##

```python
image_path=r'C:\Users\Tanvir\Downloads\field-image.jpg'
image = cv2.imread(image_path)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

## Apply Gaussian blur ##

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
```

## Perform edge detection using Canny ##

```python
canny_edges = cv2.Canny(blurred_image, 50, 150)
```

## Find contours in the edge-detected image ##

```python
contours, _ = cv2.findContours(canny_edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

## Create a copy of the original image ##

```python
crack_image = image.copy()
```

## Draw the contours on the copy of the original image ##

```python
cv2.drawContours(crack_image, contours, -1, (0, 255, 0), 2)
```

## Plot the original and crack-detected images using matplotlib ##

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(cv2.cvtColor(crack_image, cv2.COLOR_BGR2RGB))
axs[1].set_title('Crack-Detected Image')
axs[1].axis('off')

plt.show()
```
